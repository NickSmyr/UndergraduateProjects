```python
def format_cls_report(cls_report):
    entries = [list(filter(lambda x : len(x) > 0 , x.split(" "))) for x in cls_report.split("\n")][3:-5]
    return [ [int(x[0]) , float(x[1]) , float(x[2]) , float(x[3]) ,int(x[4])] for x in entries]
    
format_cls_report(class_report)
```

```python
def filter_entries(entries , f1_less_than=None , support_more_than=None):
    """
    Returns the id of the entries that have f1 less than input , and support more than input
    """
    filter1 = lambda x : (x[3] < f1_less_than if f1_less_than else True)
    filter2 = lambda x : (x[4] > support_more_than if support_more_than else True)
    return list(map(lambda x : x[0] , filter(filter2 , filter(filter1  , entries ))))
```

```python
def f1_entries(entries , f1_less_than=None , support_more_than=None):
    """
    Returns the id of the entries that have f1 less than input , and support more than input
    """
    filter1 = lambda x : (x[3] < f1_less_than if f1_less_than else True)
    filter2 = lambda x : (x[4] > support_more_than if support_more_than else True)
    return list(map(lambda x : [x[3] , x[4]] , filter(filter2 , filter(filter1  , entries ))))
```
