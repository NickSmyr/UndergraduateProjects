```python
def classification_report_formatted(report):
    formatted = [x.split(' ') for x in report.split('\n')  if len(x) > 0] 
    non_zero = [[y for y in x if len(y)> 0] for x in formatted]
    return non_zero[1:][:-3]
def labels(formatted):
    return [x[0] for x in formatted]
def f1_scores(formatted):
    return [float(x[3]) for x in formatted]
    
res_thess = classification_report_formatted(thess)
res_ours = classification_report_formatted(ours)
labels(res_ours) == labels(res_thess)
```

```python
m_scores = [f1_scores(classification_report_formatted(thess)), f1_scores(classification_report_formatted(ours))] 
m_labels = labels(res_ours)
from result_transformation import plot
plot(m_scores,['thess', 'ours'], labels=m_labels, fontsize_label=8, fontsize_per=8,figsize=(15, 6))
```
