```python
def prec(tp, fp):
    return tp/(tp+fp)
def recall(tp, fn):
    return tp/(tp+fn)
def f1(p, r):
    return (2*p*r)/(p+r)
```

```python
fn = sw.evals['macro-f1']._fn
```

```python
fp = sw.evals['macro-f1']._fp
```

```python
tp = sw.evals['macro-f1']._tp
```

```python
l = [ 'Label : %s Prec %.2f Rec %.2f F1 %.2f' % 
        (l ,prec(tp[l] ,fp[l]) , recall(tp[l] , fn[l]) , f1(prec(tp[l] ,fp[l]) , recall(tp[l] , fn[l]))) 
        for l in fn ]
```
