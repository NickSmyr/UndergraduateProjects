```python
def create_and_print_I2L(filepath):
    f = open(filepath, encoding="utf8")
    tags = []
    tagCounts = {}
    for line in f:
        tokens = line.split()
        if len(tokens) < 2:
          continue
        tag = tokens[-1]
        if tag not in tags:
          tags.append(tag)
        
        # an den uparxei vale to tag k to B/I
        if tag[2:] not in tagCounts:
          tagCounts[tag[2:]] = tag[0]
        # an uparxei check to tag
        else:
          if(tag[0] in tagCounts[tag[2:]]):
            continue 
          else:
            tagCounts[tag[2:]] = tagCounts[tag[2:]]+tag[0]
    return tags,tagCounts
```

```python
filename = 'train_elNER18.iobes'
f = os.path.join(path, filename)
tags, tagCounts = create_and_print_I2L(f)
```
