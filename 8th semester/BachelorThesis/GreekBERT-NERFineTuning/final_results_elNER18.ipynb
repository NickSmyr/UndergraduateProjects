```python
def transform_classification_report(text_scores):
    """
    Returns a strict tabular version of the classification report
    each row is in the form 
    label_id , rec , prec ,f1 , support 
    """
    temp = [ list(filter(lambda x : len(x) > 0 ,l.split(" "))) for l in text_scores.split("\n") if len(l) > 0 ]
    return [ l for l in temp if len(l) == 5 ][1:]

def get_f1_scores_per_label(text_scores):
    return [ float(l[3]) for l in transform_classification_report(text_scores)]
```
